### Definiciones

Series: Una Serie es una estructura de datos unidimensional en la biblioteca de Python llamada pandas. Es como una columna en una tabla de datos o una matriz unidimensional. Cada elemento en una serie tiene una etiqueta, llamada índice, que ayuda a identificar o acceder a los elementos en la serie. Las Series son útiles para trabajar con datos unidimensionales, como series de tiempo, listas de valores o cualquier otro tipo de datos que necesiten ser etiquetados o indexados.

Data Frame: Un DataFrame es una estructura de datos bidimensional en pandas, que es esencialmente una tabla de datos con filas y columnas.Es como una hoja de cálculo o una base de datos relacional, donde cada columna puede representar un tipo de datos diferente y cada fila contiene un conjunto de datos relacionados entre sí. Los DataFrames son extremadamente útiles para manipular datos tabulares de manera eficiente y realizar análisis de datos complejos.

Panel: Un Panel es una estructura de datos tridimensional en pandas. Mientras que las Series son unidimensionales y los DataFrames son bidimensionales, los Paneles son tridimensionales, lo que significa que pueden contener datos organizados en tres dimensiones: ítems, ejes de fila y ejes de columna. En la práctica, los Paneles son menos comunes que las Series y los DataFrames, y se utilizan principalmente en situaciones donde se necesite trabajar con datos que requieran una representación tridimensional, como datos longitudinales o datos de series temporales con múltiples variables y observaciones. Sin embargo, en muchas situaciones, los DataFrames se pueden utilizar de manera efectiva para manejar datos tridimensionales, lo que hace que el uso de Paneles sea menos necesario en la mayoría de los casos.

Construya un Panel con los tres (3) dataframes cargados anteriormente: df1, df2 y df3. Para construir este panel, cree un diccionario llamado panel que en sus keys tenga los nombres de los conjuntos de datos y los valores sean los dataframes.

### Paso 5
Programe una función que extraiga todas las listas de los encabezados de columnas que se encuentran en un Panel y úsela para extraer las listas de columnas del Panel del punto anterior. Imprima los resultados de forma que se diferencien los encabezados entre los dataframes en el Panel.

```python
def encabezados (x):
    encabezados={}
    for nombre, df in x.items():
        encabezados[nombre]=df.columns
    return encabezados
```

```python
encab=encabezados(panel)
```

```python
print("Encabezados")
print("\n")
for k, v in encab.items():
    print(k, ":")
    for i in v:
        print("\t", "-", i)
    print("\n")
      
```

### Paso 6
Programe una función que imprima la cantidad de elementos que tienen las bases de datos en un Panel y úsela con el Panel del punto anterior. Imprima los resultados de forma que se diferencien los tamaños de los diferentes dataframes en el Panel. 

```python
def dimension (x):
    dim={}
    for nombre, df in x.items():
        dim[nombre]=df.shape
    return dim
```

```python
dim = dimension(panel)
```

```python
dim
```

```python
print("Dimensiones")
print("\n")
for k, v in dim.items():
    print(k, ":") 
    print("-  Filas:", v[0])
    print("- Columnas:", v[1])
    print("\n")
 
```
