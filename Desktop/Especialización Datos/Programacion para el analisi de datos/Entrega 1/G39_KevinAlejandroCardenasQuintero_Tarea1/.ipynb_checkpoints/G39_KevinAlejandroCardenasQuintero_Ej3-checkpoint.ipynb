## Teor√≠a

Los gr√°ficos cient√≠ficos son herramientas fundamentales en la comunicaci√≥n de resultados de investigaci√≥n, ya que permiten visualizar datos de manera efectiva y comprensible. Estos gr√°ficos se utilizan en una amplia variedad de disciplinas cient√≠ficas. Estos son algunos de los gr√°ficos cient√≠ficos mas comunes y sus aplicaciones:

1. **Gr√°ficos de dispersi√≥n:**

Los gr√°ficos de dispersi√≥n son utilizados para visualizar la relaci√≥n entre dos variables continuas.
Cada punto en el gr√°fico representa una observaci√≥n y su posici√≥n est√° determinada por los valores de las dos variables.
Estos gr√°ficos son √∫tiles para identificar patrones, tendencias y correlaciones entre las variables.

2. **Histogramas:**

Los histogramas son utilizados para representar la distribuci√≥n de una variable continua.
El eje x representa los valores de la variable, mientras que el eje y representa la frecuencia de ocurrencia de cada valor.
Los histogramas son √∫tiles para identificar la forma de la distribuci√≥n, incluyendo su sesgo, simetr√≠a y presencia de picos o modas.

3. **Gr√°ficos de barras:**

Los gr√°ficos de barras son utilizados para comparar la magnitud de diferentes categor√≠as o grupos.
Cada barra representa una categor√≠a y su altura indica la magnitud de la variable que se est√° comparando.
Estos gr√°ficos son √∫tiles para visualizar diferencias entre categor√≠as y para hacer comparaciones entre grupos.

4. **Gr√°ficos de l√≠neas:**

Los gr√°ficos de l√≠neas son utilizados para representar cambios en una variable a lo largo del tiempo o en funci√≥n de otra variable.
Cada punto en el gr√°fico est√° conectado por una l√≠nea, lo que permite visualizar tendencias y patrones temporales.

5. **Diagramas de caja (Boxplots):**

Los diagramas de caja son utilizados para representar la distribuci√≥n de una variable y resumir sus estad√≠sticos clave.
Estos gr√°ficos muestran la mediana, los cuartiles y los valores at√≠picos de una distribuci√≥n, proporcionando una descripci√≥n visual de la dispersi√≥n de los datos.

6. **Diagramas de dispersi√≥n tridimensional:**

Estos gr√°ficos son una extensi√≥n de los gr√°ficos de dispersi√≥n convencionales, pero representan la relaci√≥n entre tres variables en lugar de dos.
Son √∫tiles para visualizar relaciones m√°s complejas entre m√∫ltiples variables.

7. **Mapas de calor (Heatmaps):**

Los mapas de calor son utilizados para visualizar la intensidad de una variable en una matriz bidimensional.
Los valores de la variable se representan mediante colores, lo que permite identificar patrones espaciales o de densidad en los datos.

## Gr√°fico 1:
Esta visualizaci√≥n debe cumplir con las siguientes condiciones:

‚Ä¢ Mostrar los puntos de dispersi√≥n del emparejamiento las siguientes listas:
$x= [13,14,15,16,17,18]$ , $y= [16,24,45,57,75,102]$

‚Ä¢ Mostrar la funci√≥n lineal $ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë$ en el dominio [13,18].

‚Ä¢ Mostrar la funci√≥n exponencial $ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ(ùüé.ùüëùüîùüóùíô)$ en el dominio [13,18].

‚Ä¢ Mostrar la funci√≥n logar√≠tmica $ùíâ(ùíô) = ùüêùüìùüó. ùüíùüó ùíçùíè(ùíô) ‚àí ùüîùüìùüî. ùüíùüî $en el dominio [13,18].

‚Ä¢ Los objetos mostrados deben tener colores diferentes.

‚Ä¢ La l√≠nea de la funci√≥n logar√≠tmica debe ser punteada.

‚Ä¢ En los ejes deben visualizarse los n√∫meros.

‚Ä¢ Asigne un nombre (label) a cada objeto e imprima la legenda
que los identifique en la gr√°fica.

‚Ä¢ Asigne nombres a los ejes de la gr√°fica.

‚Ä¢ Asigne un t√≠tulo a la gr√°fica.

‚Ä¢ Tenga en cuenta que los nombres que decida utilizar le ser√°n de
ayuda para presentar las conclusiones que usted obtenga del ejercicio.el ejercicio.



```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([13, 14, 15, 16, 17, 18])
y = np.array([16, 24, 45, 57, 75, 102])

# Funciones
def f(x):
    return 17*x - 210.3334

def g(x):
    return 0.1452 * np.exp(0.369*x)

def h(x):
    return 259.49 * np.log(x) - 656.44

# Gr√°fico Disperci√≥n

plt.scatter(x, y, color='red', label='Datos')

#crear valores de x en el dominio
x = np.linspace(13, 18, 100)

# Gr√°fico lineal

plt.plot(x, f(x), color='blue', label='f(x) := 17x - 210.3334')
plt.plot(x, g(x), color='green', label='g(x) := 0.1452 * e^(0.369x)')
plt.plot(x, h(x), '--', color='orange', label='h(x) := 259.49 * ln(x) - 656.44')

# Etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gr√°fico de dispersi√≥n con funciones')

# Leyenda
plt.legend()

# Mostrar gr√°fico
plt.grid(True)
plt.show()

```

### An√°lisis de resultados.

Los puntos rojos representan los datos de dispersi√≥n, que muestran la relaci√≥n entre las variables x e y, a simple vista, los datos parecen seguir una tendencia ascendente a medida que aumenta el valor de x.

El objetivo de trazar estas funciones junto con los datos es evaluar qu√© tan bien se ajustan las funciones a los datos reales. Si las funciones modelan correctamente los datos, las l√≠neas pasar√°n cerca de los puntos de dispersi√≥n.

La funci√≥n lineal en color azul parece ajustarse razonablemente bien a los puntos de dispersi√≥n en la parte inferior izquierda del gr√°fico.
Sin embargo, a medida que avanzamos hacia la parte superior derecha del gr√°fico, la discrepancia entre la funci√≥n y los datos observados se hace m√°s evidente.

La funci√≥n exponencial en color verde parece no ajustarse adecuadamente a los puntos de dispersi√≥n. Se observa una discrepancia significativa entre la funci√≥n y los datos en toda la gama de valores.

La funci√≥n logar√≠tmica en color naranja muestra un ajuste mixto a los datos observados. En la parte inferior izquierda del gr√°fico, la funci√≥n logar√≠tmica parece ajustarse razonablemente bien a los datos, pero a medida que nos movemos hacia la parte superior derecha, la discrepancia entre la funci√≥n y los datos se hace m√°s evidente.

## Gr√°fico 2:

Construya los c√≥digos necesarios para:

‚Ä¢ Generar 4 listas aleatorias de 700 muestras (revise la documentaci√≥n de numpy.random):
1. x1 con distribuci√≥n normal de media 5 y desviaci√≥n est√°ndar 0.1.
2. y1 con distribuci√≥n normal de media 7 y desviaci√≥n est√°ndar 0.1.
3. x2 con distribuci√≥n uniforme entre el valor m√≠nimo de x1 y el valor m√°ximo de x1.
4. y2 con distribuci√≥n uniforme entre el valor m√≠nimo de y1 y el valor m√°ximo de y1.

‚Ä¢ Graficar en un mismo plot nubes de dispersi√≥n de los pares de puntos que se forman con (x1, y1) y (x2, y2).

‚Ä¢ La nube de puntos (x2, y2) debe visualizarse sobre la nube de (x1, y1).

‚Ä¢ Anexe los t√≠tulos y legendas necesarios para poder citar y explicar realizar el an√°lisis del gr√°fico obtenido y las conclusiones.

```python
# Generaci√≥n de datos
np.random.seed(0)  # Para reproducibilidad

# Generar x1 e y1 con distribuci√≥n normal
x1 = np.random.normal(loc=5, scale=0.1, size=700)
y1 = np.random.normal(loc=7, scale=0.1, size=700)

# Generar x2 e y2 con distribuci√≥n uniforme
x2 = np.random.uniform(low=min(x1), high=max(x1), size=700)
y2 = np.random.uniform(low=min(y1), high=max(y1), size=700)

# Gr√°fico de dispersi√≥n
plt.scatter(x1, y1, color='blue', label='Datos (x1, y1)', alpha=0.5)
plt.scatter(x2, y2, color='red', label='Datos (x2, y2)', alpha=0.7)

# T√≠tulos y leyenda
plt.title('Nubes de dispersi√≥n de datos generados')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostrar gr√°fico
plt.grid(True)
plt.show()
```

### An√°lisis de resultados.

Se trazan dos nubes de puntos en el gr√°fico, una en color azul para los datos (x1, y1) y otra en color rojo para los datos (x2, y2).
Cada punto en el gr√°fico representa un par de coordenadas (x, y), donde x corresponde a los valores en el eje horizontal y y corresponde a los valores en el eje vertical.

Los puntos se distribuyen aleatoriamente en el gr√°fico debido a la naturaleza aleatoria de los datos generados.
No parece haber una relaci√≥n evidente entre los conjuntos de datos (x1, y1) y (x2, y2), ya que se distribuyen de manera independiente y no muestran una tendencia clara.

La nube de puntos azules (x1, y1) muestra una distribuci√≥n m√°s concentrada alrededor de un cierto rango de valores, lo que indica una menor dispersi√≥n en comparaci√≥n con la nube de puntos rojos (x2, y2).
La nube de puntos rojos (x2, y2) muestra una distribuci√≥n m√°s uniforme debido a la generaci√≥n de datos con una distribuci√≥n uniforme.

## Gr√°ficos 3:

Genere las siguientes dos visualizaciones que se proponen a continuaci√≥n:

‚Ä¢ Cargue en la variable df1 el conjunto de datos que se presentan en el archivo Excel, Anexo 1 - Base de Datos Excel.xlsx, que se encuentra en la carpeta de la Tarea 1 - Fundamentos de programaci√≥n cient√≠fica con Python.

‚Ä¢ En una misma gr√°fica presente 2 histogramas de las variables Tiempo motor ON y Tiempo en Movimiento, respectivamente. Deben ser lo suficientemente transparentes, para poder ver ambas al tiempo. (Revise la documentaci√≥n de la funci√≥n matplotlib.pyplot.hist).

‚Ä¢ En otra gr√°fica, genere el gr√°fico de dispersi√≥n que se obtiene de emparejar los valores de Tiempo motor ON y Tiempo en Movimiento.

‚Ä¢ Anexe los t√≠tulos y legendas necesarios en ambos gr√°ficos para poder citar y explicar el an√°lisis de los gr√°ficos obtenidos y las conclusiones.

```python
import pandas as pd

# Cargar los datos desde el archivo Excel
df1 = pd.read_excel('Anexo 1 - Base de Datos EXCEL.xlsx', sheet_name = "BASE DE DATOS",  header=3)

# Gr√°fico de histogramas
plt.figure(figsize=(10, 5))

# Histograma de Tiempo motor ON
plt.hist(df1['Tiempo motor ON (h)'], bins=30, alpha=0.5, color='blue', label='Tiempo motor ON', density=True)

# Histograma de Tiempo en Movimiento
plt.hist(df1['Tiempo en Movimiento (h)'], bins=30, alpha=0.5, color='green', label='Tiempo en Movimiento', density=True)

# T√≠tulos y leyendas
plt.title('Histograma de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo')
plt.ylabel('Densidad de frecuencia')
plt.legend()

# Mostrar gr√°fico
plt.grid(True)
plt.show()

# Gr√°fico de dispersi√≥n
plt.figure(figsize=(8, 6))

# Dispersi√≥n de Tiempo motor ON vs Tiempo en Movimiento
plt.scatter(df1['Tiempo motor ON (h)'], df1['Tiempo en Movimiento (h)'], color='red', alpha=0.5)

# T√≠tulos y leyendas
plt.title('Dispersi√≥n de Tiempo motor ON vs Tiempo en Movimiento')
plt.xlabel('Tiempo motor ON')
plt.ylabel('Tiempo en Movimiento')

# Mostrar gr√°fico
plt.grid(True)
plt.show()
```

### An√°lisis de resultados.

**Histogramas de Tiempo motor ON y Tiempo en Movimiento:**

-Se observa que la mayor√≠a de los datos se concentran en ciertos rangos de tie, llegando al tiempo 3mpo, lo que sugiere que hay per√≠odos en los que el motor est√° encendido y el veh√≠culo est√° en movimiento durante lapsos espec√≠ficos.

-La distribuci√≥n de tiempo en movimiento parece ser m√°s dispersa que la distribuci√≥n del tiempo del motor encendido.

**Gr√°fico de dispersi√≥n de Tiempo motor ON vs Tiempo en Movimiento**

El gr√°fico de dispersi√≥n muestra la relaci√≥n entre las variables "Tiempo motor ON" y "Tiempo en Moviento. Seento".

Se observa una tendencia general de incremento en el tiempo en movimiento a medida que aumenta el tiempo del motor encendido, lo que sugiere que hay una relaci√≥n positiva entre estas dos variables. Sin embargo, tambi√©n se observa cierta dispersi√≥n en los datos, lo que indica que hay otros factores que pueden influir en el tiempo en movimiento adem√°s del tiempo del motor encendido.
