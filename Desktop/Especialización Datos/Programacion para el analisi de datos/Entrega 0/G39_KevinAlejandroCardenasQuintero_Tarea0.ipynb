### ¿Qué es una celda markdown?

Imagina que estás escribiendo un documento en tu computadora, pero en lugar de simplemente escribir texto plano, puedes agregar formato, como negrita, cursiva, listas y encabezados, para hacer que tu documento sea más legible y estructurado. Bueno, una celda Markdown es como un bloque en tu documento donde puedes hacer exactamente eso, pero dentro de un entorno de Jupyter.

En una celda Markdown, puedes escribir texto normal como en un procesador de texto, pero también puedes usar un conjunto de reglas simples para dar formato al texto. Por ejemplo, si quieres que una palabra esté en negrita, simplemente la envuelves entre dos asteriscos . Si quieres crear una lista, solo tienes que comenzar cada elemento con un guion o un número seguido de un punto.

### ¿Qué es un paquete (o librería) en Python?

En Python, los paquetes o librerías son conjuntos de funciones y métodos que realizan tareas específicas. Por ejemplo, si necesitas trabajar con números complejos, hay un paquete llamado cmath que te proporciona todas las herramientas necesarias. Si quieres hacer gráficos bonitos, hay una librería llamada matplotlib que te facilita esa tarea.

Lo bueno de los paquetes y librerías es que no tienes que reinventar la rueda cada vez que necesitas hacer algo nuevo en Python. En lugar de eso, puedes simplemente importar el paquete o la librería que necesitas y utilizar las herramientas que proporciona. Esto te permite escribir menos código y trabajar de manera más eficiente.

### Describa para qué sirven cada una de las siguientes librerías: Pip, NumPy, Pandas, Seaborn, matplotlib, scikit-learn, Requests, Bokeh, y Folium.

- _**Pip**_:  Pip te permite instalar fácilmente paquetes de Python desde el repositorio oficial conocido como PyPI (Python Package Index). Simplemente necesitas ejecutar el comando pip install nombre_del_paquete en la línea de comandos de tu terminal, y Pip descargará e instalará el paquete especificado junto con sus dependencias. Pip también te permite actualizar los paquetes instalados a sus últimas versiones disponibles en PyPI. Esto se hace utilizando el comando pip install --upgrade nombre_del_paquete.

- _**NumPy**_: El objeto principal de NumPy es el numpy.ndarray, que es una matriz multidimensional homogénea y eficiente en términos de almacenamiento de datos. Estas matrices pueden ser de una, dos o más dimensiones y contienen elementos del mismo tipo de datos. NumPy proporciona funciones y operadores para realizar operaciones matemáticas y estadísticas en matrices de forma eficiente, sin la necesidad de utilizar bucles explícitos. Este enfoque se conoce como vectorización y es fundamental para el rendimiento y la eficiencia de NumPy.

- _**Pandas**_: El objeto principal en Pandas es el DataFrame, que es una estructura de datos tabular bidimensional similar a una hoja de cálculo. Los DataFrames constan de filas y columnas etiquetadas, lo que facilita la manipulación y el análisis de datos tabulares. Además del DataFrame, Pandas también proporciona la Serie, que es una estructura de datos unidimensional etiquetada que puede contener cualquier tipo de datos. Las Series se utilizan comúnmente para representar una sola columna o fila de un DataFrame. Adicional, permite la selección y filtrado, operaciones, analisis, visualizacion y limpieza de datos.

- _**Seaborn**_: Nos da una variedad de estilos predefinidos que mejoran la apariencia de los gráficos de forma automática. Estos estilos incluyen ajustes en colores, fuentes, tamaños y otros elementos visuales que hacen que los gráficos sean más agradables. Seaborn proporciona una amplia gama de funciones para crear gráficos estadísticos complejos con una sola línea de código. Esto incluye gráficos de dispersión con regresión lineal, gráficos de barras con intervalos de confianza, gráficos de violín, gráficos de cajas, matrices de correlación, entre otros. Seaborn está diseñado para ser fácil de usar y permite crear visualizaciones complejas con poco esfuerzo. La sintaxis de Seaborn es simple y consistente, lo que facilita su aprendizaje y uso, incluso para aquellos que no tenemos experiencia en visualización de datos.

- _**Matplotlib**_: Ofrece una amplia gama de tipos de gráficos, incluyendo gráficos de líneas, gráficos de dispersión, gráficos de barras, histogramas, gráficos de cajas, gráficos de contorno, gráficos de superficie, etc. Ademas, proporciona un control completo sobre cada aspecto del diseño y estilo de los gráficos. Esto incluye la personalización de colores, estilos de línea, tipos de marcadores, etiquetas de ejes, leyendas y títulos, lo que permite crear visualizaciones altamente personalizadas y estéticamente agradables.

- _**scikit-learn**_: Scikit-learn ofrece una amplia variedad de algoritmos de aprendizaje supervisado y no supervisado, así como herramientas para la selección de modelos y evaluación de rendimiento. Algunos de los algoritmos incluidos son: regresión lineal, regresión logística, máquinas de vectores de soporte (SVM), árboles de decisión, k-means, entre otros. Ademas, proporciona herramientas para el preprocesamiento de datos, incluyendo la normalización, estandarización, manejo de valores faltantes, codificación de variables categóricas, y extracción de características, lo que facilita la preparación de datos para el entrenamiento de modelos.

- _**Requests**_: Requests proporciona una interfaz de alto nivel que simplifica la realización de solicitudes HTTP en comparación con otras opciones disponibles en Python. Esto permite interactuar con diferentes tipos de recursos en servidores web, como recuperar datos, enviar formularios, actualizar recursos y eliminar recursos.

```python
diccionario =  {
    "Juan": (25, 70, 1.75, True),
    "María": (30, 65, 1.60, False),
    "Pedro": (40, 80, 1.80, True),
    "Ana": (20, 55, 1.70, True),
    "Carlos": (35, 75, 1.78, False),
    "Laura": (28, 60, 1.65, True),
    "Luis": (45, 85, 1.75, False),
    "Sofía": (22, 58, 1.63, True)
}
```

```python
c= diccionario["Juan"]
print(c)
```

```python
edades = [] #creacion de la listas de edades
pesos = []#creacion de la listas de pesos
estaturas = []#creacion de la listas de estaturas
estudia = []#creacion de la listas de estudia

for nom, info in diccionario.items(): #Hace un recorrido de los nombres del diccionario y los valores del diccionario
    edades.append(info[0]);# Asigna los valores del diccionario a cada lista
    pesos.append(info[1]);
    estaturas.append(info[2]);
    estudia.append(info[3]);
```

```python
def calcular_promedio(lista):

    suma = sum(lista) #suma de la lista
    promedio = suma / len(lista) #promedio de la lista
    return promedio

# Usamos la función para imprimir el promedio de las listas generadas anteriormente
print("Promedio de edades:", calcular_promedio(edades))
print("Promedio de pesos:", calcular_promedio(pesos))
print("Promedio de estaturas:", calcular_promedio(estaturas))
print("Promedio de estudia:", calcular_promedio(estudia))
```
