## Revisión Teórica: Combinación, Unión y Concatenación de DataFrames

La manipulación y combinación de conjuntos de datos son tareas fundamentales en el análisis de datos y la ciencia de datos. En el contexto de programación en Python con la biblioteca Pandas, la combinación, unión y concatenación de DataFrames son operaciones comunes para trabajar con conjuntos de datos tabulares.

### Concatenación de DataFrames

La concatenación de DataFrames implica la unión de estos a lo largo de un eje específico, ya sea a lo largo de las filas **axis=0** o a lo largo de las columnas **axis=1**. Esto permite agregar filas o columnas adicionales a un DataFrame existente. En Pandas, la función **concat()** se utiliza para realizar esta operación.

La fórmula para la concatenación de DataFrames se puede expresar de la siguiente manera:
$$
\text{concatenación}(DF_1, DF_2, \ldots, \text{axis}=0 \text{ o } 1)
$$

Donde $DF_1$, $DF_2$, son los DataFrames que se desean concatenar, y **axis** especifica el eje a lo largo del cual se realizará la concatenación.



### Unión de DataFrames

La unión de DataFrames es similar a la operación de unión en SQL, donde se combinan los conjuntos de datos en función de una o más columnas comunes. En Pandas, la función **join()** se utiliza para realizar esta operación, permitiendo especificar las columnas clave en las que se realizará la unión.

La fórmula para la unión de DataFrames se puede expresar como:
$$
\text{unión}(DF_1, DF_2, \ldots, \text{on}=\text{columna.clave})
$$

Donde $DF_1$, $DF_2$ son los DataFrames que se desean unir, y **on** especifica la(s) columna(s) clave para realizar la unión.

### Combinación de DataFrames

La combinación de DataFrames implica fusionar los conjuntos de datos en función de un índice común o de columnas específicas, similar a la unión pero sin descartar filas con valores que no tienen correspondencia en ambos conjuntos de datos. En Pandas, la función **merge()** también se usa para esta operación, con el parámetro **how** que especifica el tipo de combinación (por ejemplo, **inner**, **outer**, **left**, **right**).

La fórmula para la combinación de DataFrames se puede expresar como:
$$
\text{combinación}(DF_1, DF_2, \ldots, {how}= tipo de combinación)
$$

Donde $DF_1$, $DF_2$, son los DataFrames que se desean combinar, y **how** especifica el tipo de combinación a realizar.



## Operaciones de Agregación en Pandas

Las operaciones de agregación son fundamentales en el análisis de datos, permitiendo resumir y obtener información estadística de conjuntos de datos. En Pandas, las operaciones de agregación se aplican comúnmente a columnas o filas de DataFrames para calcular estadísticas resumidas, como la suma, el promedio, la mediana, la desviación estándar, entre otros.

Algunas de las operaciones de agregación más comunes en Pandas incluyen:

- **Suma**: Calcula la suma de los valores en una columna o fila.
- **Promedio**: Calcula el promedio de los valores en una columna o fila.
- **Mediana**: Calcula la mediana de los valores en una columna o fila.
- **Desviación estándar**: Calcula la desviación estándar de los valores en una columna o fila.
- **Mínimo y máximo**: Encuentra el valor mínimo y máximo en una columna o fila.

En Pandas, las operaciones de agregación se aplican utilizando métodos como `sum()`, `mean()`, `median()`, `std()`, `min()`, `max()`, entre otros. Estos métodos pueden aplicarse directamente a DataFrames o Series, y pueden ser aplicados a través de columnas específicas o a todo el DataFrame.

```python
import pandas as pd
from google.colab import drive


# Creamos un DataFrame de ejemplo 'df'
df = pd.read_csv("/content/drive/MyDrive/Anexo 1 - Student Weight Status.csv")

# Creamos p1 con los datos del 0 al 1 de 'df'
p1 = df[["AREA NAME", "COUNTY"]][0:2] # se debe utilizar un numero mayor al de las filas df

# Creamos p2 con los datos del 2 al 4 de 'df'
p2 = df[["AREA NAME", "COUNTY"]][2:5]

# Concatenamos p1 y p2 y mostramos los resultados Colocando el p1 arriba y el p2 abajo
concatenado_simple = pd.concat([p1, p2])

concatenado_simple

```

```python
# Concatenamos p1 y p2 identificando las asignaciones por las llaves 'p1' y 'p2'
concatenado_con_llaves = pd.concat([p1, p2], keys=['p1', 'p2'])

concatenado_con_llaves
```

```python
# Rescatamos los datos de p1 desde el DataFrame 'concatenado_con_llaves'
datos_p1 = concatenado_con_llaves.loc['p1'] #Trae los valores solo de p1

datos_p1
```

La diferencia entre la concatenación simple y la concatenación con identificación por llaves es que en la segunda opción, se agrega un nivel de índice adicional que identifica de dónde provienen los datos (p1 y p2 en este caso).

Al rescatar los datos de p1 desde el DataFrame concatenado, obtenemos una subselección del DataFrame original p1 con la misma estructura y valores.
