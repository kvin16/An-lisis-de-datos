## Teoria

**Limpieza y preprocesamiento de datos:**

La limpieza y el preprocesamiento de datos son etapas fundamentales en cualquier proyecto de análisis de datos. Consisten en identificar y corregir problemas en los datos, como valores faltantes, incorrectos o inconsistentes. Este proceso es crucial para garantizar la calidad de los datos y obtener resultados precisos y confiables en el análisis posterior.

**Segmentación y filtrado de datos:**

La segmentación y el filtrado de datos implican seleccionar subconjuntos específicos de datos basados en ciertos criterios o condiciones. Esto permite a los analistas centrarse en la información relevante para sus objetivos de análisis y explorar conjuntos de datos grandes y complejos de manera más efectiva.

**Normalización de datos:**

La normalización de datos se refiere al proceso de convertir los datos en un formato estándar y uniforme. Esto puede implicar convertir cadenas de texto en mayúsculas o minúsculas, estandarizar formatos de fechas o números, y eliminar caracteres especiales o espacios en blanco innecesarios. La normalización facilita la comparación y el análisis de los datos, así como evita errores debidos a diferencias en el formato.

## Paso 1. Preguntas:

Para este ejercicio, cargue de nuevo los datos del Ejercicio 1, pero genere un DataFrame llamado df, solamente con las columnas:
- type
- mass_kg
- year_built • home_port • name

```python
import requests
import pandas as pd

def obtener_datos():
    url = "https://api.spacexdata.com/v4/ships" #Formato String

    try:
        response = requests.get(url) #Trae los datos de la url
        response.raise_for_status()  # Si hay un error en la solicitud, lanzará una excepción
        datos_naves = response.json()  # Convertir la respuesta a formato JSON
        return datos_naves
    except requests.exceptions.RequestException as e:
        print("Error al realizar la solicitud:", e)
        return None

datos_naves = obtener_datos()

df = pd.DataFrame.from_dict(datos_naves)

df = df[["type", "mass_kg", "year_built", "home_port", "name"]] # Se seleccionan las columnas deseadas en el dataframe df
```

## Paso 3

En el DataFrame df completa los datos faltantes por los siguientes tres métodos (muestra y comenta los resultados):

  * Completar datos faltantes con el número 0.
  *  Completar por método de propagación hacia adelante.
  * Completar los valores faltantes con el promedio de la
columna.

```python
# Completar datos faltantes con el número 0
df_zero = df.fillna(0) #filna indica con el valor que tu necesites reemplzar, en este caso 0

# Completar por método de propagación hacia adelante
df_propagacion = df.ffill() #Toma el valor anterior no nulo y con este valor completa el dato

# Completar los valores faltantes con el promedio de la columna
df_promedio = df.copy()
df_promedio.mass_kg = df.mass_kg.fillna(df.mass_kg.mean()) #filna indica con el valor que tu necesites reemplzar, en este caso el promedio de masa kg
df_promedio.year_built = df.year_built.fillna(df.year_built.mean()).astype(int) #filna indica con el valor que tu necesites reemplzar, en este caso el promedio del año y colocandolo entero
df_promedio
```
